### Example 5: Supporting new models and languages


The main components of **Microsoft Presidio** that process text and must be adapted when adding support for a new language are:

1. **NLP Engine (`NlpEngine`)**
   - Contains the underlying NLP model.
   - Responsible for core NLP tasks, including:
     - Tokenization
     - Lemmatization
     - Named Entity Recognition (NER)
     - Other language-specific linguistic processing

2. **PII Recognizers (`EntityRecognizer`)**
   - A collection of recognizers responsible for detecting specific types of Personally Identifiable Information (PII).
   - Existing recognizers may need to be **adapted** for the new language, or **new recognizers** may need to be created when language-specific patterns or entities are not covered.

```python
from presidio_analyzer import AnalyzerEngine
from presidio_analyzer.nlp_engine import NlpEngineProvider

# import spacy
# spacy.cli.download("es_core_news_md")

# Create configuration containing engine name and models
configuration = {
    "nlp_engine_name": "spacy", # the nlp engine to use
    "models": [
        {"lang_code": "es", "model_name": "es_core_news_md"},# spanish model
        {"lang_code": "en", "model_name": "en_core_web_lg"}, # enlish model
    ],
}

# Create NLP engine based on configuration
provider = NlpEngineProvider(nlp_configuration=configuration)
nlp_engine_with_spanish = provider.create_engine()

# Pass the created NLP engine and supported_languages to the AnalyzerEngine
analyzer = AnalyzerEngine(
    nlp_engine=nlp_engine_with_spanish, supported_languages=["en", "es"]
)

# Analyze in different languages

#testing the spanish model
results_spanish = analyzer.analyze(text="Mi nombre es Morris", language="es")
print("Results from Spanish request:")
print(results_spanish)

#testing the english model
results_english = analyzer.analyze(text="My name is Morris", language="en")
print("Results from English request:")
print(results_english)
```
